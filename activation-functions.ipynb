**SIGMOID FUNCTION y=1/(1+e^(-x)).**

```python
import math

def sigmoid(x):
    return 1/(1+ math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(-56)
```

**tan h**

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-56), tanh(1),tanh(50)
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(-100)
```

```python
relu(8)
```

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
