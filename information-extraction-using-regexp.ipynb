```python
import re
```

```python
text='''
Born	Elon Reeve Musk
June 28, 1971 (age 50)
Pretoria, Transvaal, South Africa
Citizenship	
South Africa (1971–present)
Canada (1971–present)
United States (2002–present)
Education	University of Pennsylvania (BS, BA)
Title	
Founder, CEO and Chief Engineer of SpaceX
CEO and product architect of Tesla, Inc.
Founder of The Boring Company and X.com (now part of PayPal)
Co-founder of Neuralink, OpenAI, and Zip2
Spouse(s)	
Justine Wilson
​
​(m. 2000; div. 2008)​
Talulah Riley
​
​(m. 2010; div. 2012)​
​
​(m. 2013; div. 2016)
'''
```

**Extracting the age.**

```python
patterns="age (\d+)"
matches=re.findall(patterns,text)
matches
```

**Extracting the name.**

```python
patterns="Born(.*)"
matches=re.findall(patterns,text)
matches[0].strip()
```

**Getting the DOB**

```python
patterns="Born.*\n(.*)\("
matches=re.findall(patterns,text)
matches[0].strip()
```

**Getting the place of birth**

```python
patterns="\(age.*\n(.*)"
matches=re.findall(patterns,text)
matches
```

```python
def get_pattern_matches(patterns,matches):
    matches=re.findall(patterns,text)
    if matches:
        return matches[0]
    
```

```python
get_pattern_matches('\(age.*\n(.*)',text)
```

```python
def get_personal_information(text):
    age=get_pattern_matches('age (\d+)',text)
    name=get_pattern_matches('Born(.*)',text)
    dob=get_pattern_matches('Born.*\n(.*)\(',text)
    pob=get_pattern_matches('\(age.*\n(.*)',text)
                            
    return {
        'age':int(age),
        'name':name.strip(),
        'dob':dob.strip(),
        'pob':pob.strip()
    }

    
```

```python
get_personal_information(text)
```
