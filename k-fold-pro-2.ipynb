```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
li=load_iris()
dir(li)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
svc_score=(cross_val_score(SVC(),li.data,li.target))
np.average(svc_score)
```

```python
lr_score=(cross_val_score(LogisticRegression(),li.data,li.target))
np.average(lr_score)
```

```python
rfc_score=(cross_val_score(RandomForestClassifier(),li.data,li.target))
np.average(rfc_score)
```

```python
rfc_score=(cross_val_score(DecisionTreeClassifier(),li.data,li.target))
np.average(rfc_score)
```
