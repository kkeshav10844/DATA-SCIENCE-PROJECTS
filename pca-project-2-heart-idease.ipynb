```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/kaggle/input/heart-failure-prediction/heart.csv")
df.head()
```

```python
df["zscore"]=(df["MaxHR"]-df["MaxHR"].mean())/df["MaxHR"].std()
```

```python
df_no_outliers=df[(df["zscore"]<3) & (df["zscore"]>-3)]
```
