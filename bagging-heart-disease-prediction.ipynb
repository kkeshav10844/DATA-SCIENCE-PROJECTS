```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/19_Bagging/Exercise/heart.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df=df[(df.Age>(df.Age.mean()-3*df.Age.std())) & (df.Age<df.Age.mean()+3*df.Age.std())]
df=df[(df.RestingBP>(df.RestingBP.mean()-3*df.RestingBP.std())) & (df.RestingBP<(df.RestingBP.mean()+3*df.RestingBP.std()))]
df=df[(df.Cholesterol>(df.Cholesterol.mean()-3*df.Cholesterol.std())) & (df.Cholesterol<(df.Cholesterol.mean()+3*df.Cholesterol.std()))]
df=df[(df.MaxHR>(df.MaxHR.mean()-3*df.MaxHR.std())) & (df.MaxHR<(df.MaxHR.mean()+3*df.MaxHR.std()))]
df=df[(df.Oldpeak>(df.Oldpeak.mean()-3*df.Oldpeak.std())) & (df.Oldpeak<(df.Oldpeak.mean()+3*df.Oldpeak.std()))]
df=df[(df.HeartDisease>(df.HeartDisease.mean()-3*df.HeartDisease.std())) & (df.HeartDisease<(df.HeartDisease.mean()+3*df.HeartDisease.std()))]
```

```python
df_c = df.copy()
df_c.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df_c.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df_c.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df_c.head()
```

```python
df_ohe=pd.get_dummies(df_c,drop_first=True)
df_ohe
```

```python
X=df_ohe.drop("HeartDisease",axis='columns')
y=df_ohe.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_scaled=scaler.fit_transform(X)
df_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df_scaled,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores=cross_val_score(SVC(),X,y,cv=5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model= BaggingClassifier(estimator=SVC(),n_estimators=100,max_samples=0.8,random_state=0)
scores=cross_val_score(bag_model,X,y,cv=5)
scores.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
scores=cross_val_score(DecisionTreeClassifier(random_state=0),X,y,cv=5)
scores.mean()
```
