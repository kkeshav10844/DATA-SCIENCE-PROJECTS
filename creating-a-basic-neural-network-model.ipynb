```python
# Importing necessary libraries.
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
```

```python
#Creating a model clas that iherits nn.module
class Model(nn.Module):
    def __init__(self,in_features,h1,h2,out_features):
        super().__init__()
        self.fc1=nn.Linear(in_features,h1)
        self.fc2=nn.Linear(h1,h2)
        self.out=nn.Linear(h2,out_features)
        
    def forward(self,x):
        x=F.relu(self.fc1(x))
        x=F.relu(self.fc2(x))
        x=self.out(x)
        
        return x
torch.manual_seed(41)
model=Model(4,8,9,3)
```

```python
#Load Data and Train Neaural Network Model.
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
url="https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv"
my_df=pd.read_csv(url)

my_df['species']=my_df['species'].replace('setosa',0.0)
my_df['species']=my_df['species'].replace('versicolor',1.0)
my_df['species']=my_df['species'].replace('virginica',2.0)


```

```python
X=my_df.drop('species',axis=1)
y=my_df['species']
X
```

```python
#set the criterion of model to measure the error, how far off the predictions are from data.
Criterion=nn.CrossEntropyLoss()
```

```python
# Choose adam optimizer, lr=learning rate(if error doesnt go down after a bunch of iterations(epochs), lower our learning rate.
Optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
epochs=100
losses=[]
for i in range(epochs):
    y_pred=model.forward(X_train)
    loss=Criterion(y_pred,y_train)
    losses.append(loss.detach().numpy())
    
    if i%10==0:
        print(f'Epoch: {i} and loss:{loss}')
```
