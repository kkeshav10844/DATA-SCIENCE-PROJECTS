```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv")
df.head()
```

```python
plt.xlabel("area(sq. ft)")
plt.ylabel("price(US $)")
plt.scatter(df.area,df.price,color="red",marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)
```

```python
df1=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/areas.csv")
```

```python
p=reg.predict(df1)
```

```python
df["prices"]=p
```

```python
plt.xlabel("area(sq. ft)")
plt.ylabel("price(US $)")
plt.scatter(df.area,df.price,color="red",marker="+")
plt.plot(df.area,reg.fit(df[["area"]]),color="blue")
```
